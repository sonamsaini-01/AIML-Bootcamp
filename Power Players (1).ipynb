```
The 6 Stages of Web Scraping
1. Sending the HTTP Request
- The scraper acts as a digital client, dispatching an HTTP GET request to the target web server’s URL.

2. Retrieving Web Page Content
The server captures the request and responds by delivering the raw code of the webpage, most commonly in HTML format.

3.Parsing the HTML Content
The unstructured raw HTML is fed into dedicated parsing tools (such as BeautifulSoup or Scrapy) to translate the code into a structured, navigable tree of elements.

4.Extracting the Data
Searches the parsed tree using predefined rules or locators to grab only the specific data points you want (e.g., product prices, headlines).

5.Storing the Data
The captured data is cleaned and exported into a permanent, structured format for analysis—typically CSV, JSON, or directly into a Database.

6.Anti-Scraping Defenses: Bypassing IP blocks, rate-limits, or CAPTCHAs via rotating proxies and headers.
Dynamic Content: Employing "headless browsers" (like Selenium or Playwright) to force JavaScript to load data that isn’t present in the initial raw HTML.
Legal & Ethical Compliance: Abiding by the site's robots.txt rules, Terms of Service, and regional data privacy laws.
    ```

````python
products = soup.find_all("div", attrs={"data-component-type": "s-search-result"})
for product in products:
    title = product.select_one("h2 span")
    if title:
        print(title.text.strip())
print(products[0].prettify())
````

````python
for laptop in laptops:
    price = laptop.find("span", class_="a-price-whole")

    if price:
        print(price.text.strip())
````

````python
for laptop in laptops:
    rating = laptop.find("span", class_="a-icon-alt")

    if rating:
        print(rating.text.strip())
````

````python
for laptop in laptops:
    img = laptop.find("img")

    if img:
        print(img["src"])
````

````python
import pandas as pd

data = []

for laptop in laptops:
    title = laptop.find("h2")
    title = title.get_text(strip=True) if title else "N/A"

    price = laptop.find("span", class_="a-price-whole")
    price = price.get_text(strip=True) if price else "N/A"

    rating = laptop.find("span", class_="a-icon-alt")
    rating = rating.get_text(strip=True) if rating else "N/A"

    img = laptop.find("img")
    image_url = img.get("src") if img else "N/A"

    data.append({
        "Title": title,
        "Price": price,
        "Rating": rating,
        "Image URL": image_url
    })

df = pd.DataFrame(data)

print(df.head())
````
