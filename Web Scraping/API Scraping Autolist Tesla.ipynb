```python
import requests
import pandas as pd

url = "https://www.autolist.com/api/v3/search"

params = {
    "ads": "true",
    "include_total_price_change": "true",
    "include_time_on_market": "true",
    "include_relative_price_difference": "true",
    "facet_counts": "true",
    "search_engine": "cgsp",
    "latitude": "37.7749295",
    "longitude": "-122.4194155",
    "limit": 20,
    "makes[]": "Tesla",
    "page": 1,
    "radius": 100
}

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/149.0.0.0 Safari/537.36",
    "Referer": "https://www.autolist.com/listings",
    "sec-ch-ua-platform": '"Windows"',
    "sec-ch-ua": '"Brave";v="149", "Chromium";v="149", "Not:A-Brand";v="24"',
    "sec-ch-ua-mobile": "?0",
    "x-autolist-client-platform": "web",
    "x-autolist-device-type": "desktop",
    "Accept": "application/json"
}

response = requests.get(url, params=params, headers=headers)

print(response.status_code)
print(response.text[:500])
```

<div style="
    background: #FADADD;
    color: #6D597A;

    border-radius: 10px;

    🌐 Step : Create json object
               Output Keys  🌐
    
</div>

```python
data = response.json()

print(data.keys())
```

<div style="
    background: #FADADD;
    color: #6D597A;
    padding: 12px;
    border-radius: 10px;
    text-align: center;
    font-size: 24px;
    font-weight: bold;
    border: 2px solid #E8C7D1;">
    
    🌐 Step : Find length of data 🌐
    
</div>

```python
records = data.get("records", [])

print("Cars:", len(records))
```

<div style="
    background: #FADADD;
    color: #6D597A;
    padding: 12px;
    border-radius: 10px;
    text-align: center;
    font-size: 24px;
    font-weight: bold;
    border: 2px solid #E8C7D1;">
    
    🌐 Step : Append data inside a list 
               Make Pandas Dataframe 🌐
    
</div>

```python
cars = []

for car in records:
    cars.append({
        "Year": car.get("year"),
        "Price": car.get("price"),
        "Mileage": car.get("mileage"),
        "City": car.get("city"),
        "State": car.get("state"),
        "VIN": car.get("vin")
    })

df = pd.DataFrame(cars)

df.head()
```
