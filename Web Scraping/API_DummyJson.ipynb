```python
import requests
import pandas as pd
import json
```

```python
url = "https://dummyjson.com/products"
response = requests.get(url)
print("Status Code:", response.status_code)
```

```python
data = response.json()["products"]   # Products are inside the "products" key
df = pd.DataFrame(data)

df.head()
```
