```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

options = Options()
# options.add_argument("--headless=new")   # Uncomment if you want headless mode

driver = webdriver.Chrome(options=options)
driver.maximize_window()

driver.get("https://www.imdb.com/chart/top/")

wait = WebDriverWait(driver, 20)

# Wait until movie titles are loaded
wait.until(
    EC.presence_of_element_located((By.CSS_SELECTOR, "h4.ipc-title__text"))
)

movies = []

# Each movie card
cards = driver.find_elements(By.CSS_SELECTOR, "li")

for card in cards:

    try:
        title = card.find_element(By.CSS_SELECTOR, "h4.ipc-title__text").text
    except:
        continue

    try:
        url = card.find_element(
            By.CSS_SELECTOR,
            "a.ipc-title-link-wrapper"
        ).get_attribute("href")
    except:
        url = ""

    try:
        metadata = card.find_elements(
            By.CSS_SELECTOR,
            "ul.ipc-inline-list li"
        )

        year = metadata[0].text if len(metadata) > 0 else ""
        runtime = metadata[1].text if len(metadata) > 1 else ""
        certificate = metadata[2].text if len(metadata) > 2 else ""

    except:
        year = runtime = certificate = ""

    try:
        rating = card.find_element(
            By.CSS_SELECTOR,
            '[data-testid="ratingGroup--imdb-rating"]'
        ).text
    except:
        rating = ""

    rank = title.split()[0].replace("#", "")
    movie_name = " ".join(title.split()[1:])

    movies.append({
        "Rank": rank,
        "Title": movie_name,
        "Year": year,
        "Runtime": runtime,
        "Certificate": certificate,
        "Rating": rating,
        "URL": url
    })

driver.quit()

df = pd.DataFrame(movies)

print(df.head())
print("Total Movies:", len(df))
```

```python
df.to_csv("imdb_top250.csv", index=False)
print("CSV saved as imdb_top250.csv")
```
