```python
import requests
```

```python
url = "https://www.goodreads.com/quotes"
headers={ "User-Agent": "Mozilla/5.0(Windows NT 10.0; Win64 ; x64) appleWebKit/537.36 (KHTML ,Like Gecko Chrome/91.0.4472.124 Safari/537.36"}
response= requests.get(url , headers=headers)
response.status_code
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
soup = BeautifulSoup(html_code, "html.parser")
#soup
```

<div style="
    background: #FADADD;
    color: #6D597A;
    padding: 12px;
    border-radius: 10px;
    text-align: center;

    🌐 Step 4 : Extracting Data 🌐
    
</div>

```python
products = soup.find_all("div", class_="quoteDetails")
```

```python
products
```

```python
data = []

for product in products:

    # Title
    quote = product.find(class_="quoteText")
    quote = quote.get_text(strip=True) if quote else "N/A"

    # Price
    author = product.find("span", class_="authorOrTitle")
    author = author.get_text(strip=True) if author else "N/A"
    
    #img
    img = product.find("img")
    image_url = img.get("src") if img else "N/A"

    data.append({
        "quote": quote,
        "author": author,
        "Image URL": image_url
    })
```

```python
df = pd.DataFrame(data)

print(df.head())
```
