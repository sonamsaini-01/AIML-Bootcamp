Web Scrapping is a powerful technique used to extract data from websites 
- fetch web pages -> parsing the content -> mark the content to be reterived -> store it for further analysis
- commonly used for-> data collection , research and various other application

How web scraping is done?
  - Sending http request to target web server to reterieve desired webpage
  - Retrieving web page content : Server responds with a web page content usually html format
  - Parsing the html content : locate specific data elements using libraries or tools designed for this purpose [BeautifulSoup, Scrapy]
  - Extracting Data : RElevant data is extracted from the parse html structure based on predefined rules or creteria
  - Storing data : Extracted data can be stored in various format as csv, json or databases for further analysis or processing.
  - Handling Challenges: Web scrappers may need to address challenges like:
            - antiscrapping measures  
            - dynamic content loading(javascript)
            - legal considerations

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Parsing the html file content stored in html_doc
soup = BeautifulSoup( html_doc, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
books=soup.find_all('div', class_ ='book')
books
```

```python
book_data=[]
```

```python
for book in books:
    title = book.find('p', class_='title').get_text(strip = True)
    author = book.find('p' , class_='author').get_text(strip=True)
    price =book.find('p' , class_='price').get_text(strip=True)
    publication_year =book.find('p' , class_='publication-year').get_text(strip=True)
    genre= book.find('p' , class_='genre').get_text(strip=True)
    description = book.find('p' , class_='description').get_text(strip=True)
    book_data.append({'Title': title,
                      'Author':author,
                      'Price' : price,
                      'Publication Year' : publication_year,
                      'Genre' : genre,
                      'Description': description
                     })

book_data
```

```python
df= pd.DataFrame(book_data)
df
```

```python
df.to_csv("Web_Scrapping_data", index = False )
```
