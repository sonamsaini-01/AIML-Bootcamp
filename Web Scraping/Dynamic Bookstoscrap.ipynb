```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd

driver = webdriver.Chrome()

driver.get("https://books.toscrape.com/")

books = []

while len(books) < 60:

    cards = driver.find_elements(By.CSS_SELECTOR, "article.product_pod")

    for card in cards:

        if len(books) == 60:
            break

        title = card.find_element(By.TAG_NAME, "h3") \
                    .find_element(By.TAG_NAME, "a") \
                    .get_attribute("title")

        price = card.find_element(By.CLASS_NAME, "price_color").text

        rating = card.find_element(By.CSS_SELECTOR, "p.star-rating") \
                     .get_attribute("class").split()[-1]

        books.append({
            "Title": title,
            "Price": price,
            "Rating": rating
        })

    # Go to the next page only if we still need more books
    if len(books) < 60:
        next_btn = driver.find_element(By.CSS_SELECTOR, "li.next a")
        next_btn.click()

driver.quit()

df = pd.DataFrame(books)

print(df)
print("Total Books:", len(df))
```

```python
df.to_csv("dynamic_bookstoscrap.csv", index=False)
print("Dynamic Books to scrap.csv successfully saved")
```
