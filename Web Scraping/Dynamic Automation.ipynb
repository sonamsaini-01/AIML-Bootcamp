```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

driver = webdriver.Chrome()

driver.get("https://automationexercise.com/products")

wait = WebDriverWait(driver, 10)

# Wait until the product cards appear
wait.until(
    EC.presence_of_element_located((By.CLASS_NAME, "product-image-wrapper"))
)

products = []

cards = driver.find_elements(By.CLASS_NAME, "product-image-wrapper")

for card in cards:
    try:
        name = card.find_element(By.CSS_SELECTOR, "div.productinfo.text-center p").text
        price = card.find_element(By.TAG_NAME, "h2").text
        link = card.find_element(By.PARTIAL_LINK_TEXT, "View Product").get_attribute("href")

        products.append({
            "Product": name,
            "Price": price,
            "URL": link
        })
    except:
        pass

driver.quit()

df = pd.DataFrame(products)

print(df)
print("Total Products:", len(df))
```

```python
df.to_csv("dynamic_automation.csv", index=False)
print("CSV saved as dynamic automation.csv")
```
