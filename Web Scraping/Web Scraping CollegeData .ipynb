```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://www.collegedata.com/college-search")

# Wait for page to load
time.sleep(5)

# Scroll until at least 100 cards are loaded
while True:
    cards = driver.find_elements(By.CSS_SELECTOR, "div[class*='CollegeCard_infoWrapper']")

    print("Loaded:", len(cards))

    if len(cards) >= 100:
        break

    # Scroll to bottom
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(3)

html = driver.page_source
driver.quit()

# Parse after scrolling
soup = BeautifulSoup(html, "html.parser")

cards = soup.select("div[class*='CollegeCard_infoWrapper']")
print("Total cards:", len(cards))
```

```python
all_colleges = []

for card in cards:
    stats = {}

    # College name
    name = card.select_one("h3")
    if name:
        stats["College Name"] = name.get_text(strip=True)

    # All label-value pairs
    for i in card.select("div[class*='StatLine_container']"):
        label = i.select_one("div[class*='StatLine_label']").get_text(strip=True).replace(":", "")
        value = i.select_one("div[class*='StatLine_value']").get_text(strip=True)
        stats[label] = value

    all_colleges.append(stats)
```

```python
all_colleges
```

```python
df = pd.DataFrame(all_colleges)

print(df.head())
```
