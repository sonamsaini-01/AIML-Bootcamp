# Warm-Up Questions

## 1. What is the difference between a list and a tuple, beyond "one is mutable and one isn't"? Why would you ever choose a tuple?

- Lists are intended for collections that may change over time.
- Tuples are intended for fixed collections of related values.
- Tuples use less memory and can be slightly faster than lists.
- Tuples can be used as dictionary keys and set elements if all their contents are hashable.
- Using a tuple signals that the data should not be modified.

**Example:** Coordinates `(10, 20)` are naturally represented as a tuple because they are fixed values.

---

## 2. Why is dictionary lookup O(1) on average? What data structure makes this possible internally?

Python dictionaries use a **hash table** internally.

When a key is inserted:
1. Python computes the hash value of the key.
2. The hash determines where the key-value pair is stored.
3. Looking up the key computes the same hash and directly accesses the location.

Therefore, dictionary lookup is **O(1) on average** because it does not need to search through all elements.

---

## 3. What does "hashable" mean, and which built-in types are NOT hashable?

A value is **hashable** if:
- It has a fixed hash value during its lifetime.
- It can be used as a dictionary key or set element.

### Hashable Types
- `int`
- `float`
- `str`
- `bool`
- `tuple` (if all elements are hashable)

### Not Hashable
- `list`
- `dict`
- `set`

These types are mutable, so their contents can change after creation.

## 5. Why is set membership testing faster than list membership testing?

### List Membership

```python
x in my_list
```

- Searches element by element.
- Time complexity: **O(n)**.

### Set Membership

```python
x in my_set
```

- Uses hashing.
- Time complexity: **O(1)** on average.

Sets are faster because they can directly locate an element instead of scanning the entire collection.

## 7. What is the difference between a list comprehension and a generator expression—both syntactically and behaviourally?

### List Comprehension - locates all memory at once 

```python
squares = [x*x for x in range(5)]
```

### Generator Expression - locates memory when called

```python
squares = (x*x for x in range(5))
```

### Differences

| List Comprehension | Generator Expression |
|-------------------|---------------------|
| Creates all values immediately | Produces values lazily |
| Uses more memory | Uses less memory |
| Supports indexing | Does not support indexing |
| Faster for small datasets | Better for large datasets |

Generators are useful when working with large amounts of data because values are generated only when needed.

### WHY this happens:

- You are **modifying the list while iterating over it**
- Python iterator does NOT re-evaluate skipped indices after removal

Step behavior:
- Start: `[1, 2, 3, 4, 5]`
- `1` → keep
- `2` → remove → list becomes `[1, 3, 4, 5]`
- iterator moves forward → **skips checking 3 correctly? (index shifts happen)**
- `3` → kept
- `4` → remove → list becomes `[1, 3, 5]`
- `5` → kept

So some elements are skipped due to shifting indices.

👉 Key idea: **Never modify a list while iterating over it**


````python
my_list = [1, 2, 3, 4, 5]
for item in my_list:
    if item % 2 == 0:
        my_list.remove(item)

print(my_list)
````

##  Bug 1 — Unexpected List Mutation

###  Problem
```python
def apply_discount_tag(orders):
    tagged = orders
    tagged.append("DISCOUNT_APPLIED")
    return tagged
 Issue:
tagged = orders → both variables point to the same list

##  Bug 5 — N+1 Lookup Problem (Slow Code)

###  Problem
#### Nested loop → O(n * m)  Fix (use dictionary lookup → O(n + m)):

````python
orders = [
    {"order_id": 1, "customer_id": 101},
    {"order_id": 2, "customer_id": 102}
]

customers = [
    {"id": 101, "name": "Asha"},
    {"id": 102, "name": "Ravi"}
]

def attach_customer_name(orders, customers):
    customer_map = {c["id"]: c["name"] for c in customers}

    for order in orders:
        order["customer_name"] = customer_map.get(order["customer_id"])

    return orders

print(attach_customer_name(orders, customers))
````

##  Bug 6 — Unhashable Type Error

###  Problem
#### t["category"] is a list → unhashable
####  Dict keys must be immutable

````python
#Fix
transactions = [
    {"date": "2026-01-01", "category": ["food"], "amount": 200},
    {"date": "2026-01-01", "category": ["food"], "amount": 150},
]

totals = {}

for t in transactions:
    key = (t["date"], tuple(t["category"]))  # convert list → tuple
    totals[key] = totals.get(key, 0) + t["amount"]

print(totals)
````

## 3. Deeply Nested Structures (SAFE ACCESSOR)

````python
def safe_get(d, *keys):
    for key in keys:
        if isinstance(d, dict) and key in d:
            d = d[key]
        else:
            return None
    return d
data = {"user": {"profile": {"address": {"city": "Pune"}}}}

print(safe_get(data, "user", "profile", "address", "city"))
print(safe_get(data, "user", "profile", "phone"))
````

# Think like an Engineer

# Failure Analysis 

### 1. Mutable default argument (cache={})
```
Using a mutable default argument means the same dictionary is reused across function calls. 
Over time, it accumulates data, causing unintended state retention and memory growth, which appears like a memory leak in production.
```
---

### 2. Deleting dict while iterating
```
Modifying a dictionary during iteration changes its size mid-loop, causing `RuntimeError` or skipping keys unpredictably.
Python does not allow structural changes during iteration because the iterator becomes invalid or inconsistent.
```
---

### 3. Using list for "seen" checks
```
List membership is O(n), so checking `uid in seen_list` becomes extremely slow at scale.
With millions of log lines, this leads to quadratic time complexity, drastically increasing runtime from minutes to hours.
```
---

### 4. Dict as default argument
```
A default dict is created once at function definition time and reused.
This means all calls share the same object, causing state leakage between calls and unpredictable configuration behavior across function executions.
```
---

## Hidden Assumptions

### 1. "I'll just use a list — order doesn't matter"
```Assumes only simplicity matters. In reality, performance matters; choosing a list ignores better structures like sets or dicts when fast lookup is needed.```

---

## 2. "Dicts don't preserve order"
``` Incorrect assumption. Modern Python preserves insertion order in dicts, so manual sorting is often unnecessary unless specific ordering is required.
```
---

## 3. "Sets and lists support same operations"
```False equivalence. Sets support fast membership and uniqueness but lack indexing and ordering, unlike lists which are ordered and allow duplicates.
```
---

## 4. "Deep copy is always safer"
```
Overuse of `deepcopy` is inefficient. It is only needed for nested mutable structures; otherwise, it adds unnecessary performance and memory overhead compared to shallow copy.```

````python
def top_scorer(names, scores):
    student_scores = dict(zip(names, scores))
    return max(student_scores, key=student_scores.get)

print(top_scorer(["Asha", "Ravi", "Meena"], [88, 95, 91]))
````

## 6. Group Words by First Letter

````python
from collections import defaultdict

def group_by_first_letter(words):
    groups = defaultdict(list)

    for word in words:
        groups[word[0]].append(word)

    return dict(groups)

print(group_by_first_letter(["apple", "ant", "banana", "cat", "car"]))
````

## 7. Find Pairs with Target Sum in O(n)

````python
def find_pairs_with_sum(numbers, target):
    seen = set()
    pairs = []

    for num in numbers:
        complement = target - num

        if complement in seen:
            pairs.append((complement, num))

        seen.add(num)

    return pairs

print(find_pairs_with_sum([2, 7, 11, 15, -2, 9], 9))
````

warehouse_b = {"SKU1": 100, "SKU2": 45, "SKU4": 10}

# 1. SKUs present in only one warehouse
only_in_a = set(warehouse_a) - set(warehouse_b)
only_in_b = set(warehouse_b) - set(warehouse_a)

print("Only in Warehouse A:", only_in_a)
print("Only in Warehouse B:", only_in_b)

# 2. SKUs present in both but with different counts
differences = {}

for sku in set(warehouse_a) & set(warehouse_b):
    if warehouse_a[sku] != warehouse_b[sku]:
        differences[sku] = warehouse_a[sku] - warehouse_b[sku]

print("Count Differences:", differences)
```

# Task 2 — Customer Segmentation

```python
customers = [
    {"name": "Asha", "total_spent": 15000},
    {"name": "Ravi", "total_spent": 500},
    {"name": "Meena", "total_spent": 4500},
    {"name": "Karan", "total_spent": 12000},
]

segments = {
    "high_value": [],
    "medium_value": [],
    "low_value": []
}

for customer in customers:
    spent = customer["total_spent"]

    if spent > 10000:
        segments["high_value"].append(customer["name"])
    elif spent >= 1000:
        segments["medium_value"].append(customer["name"])
    else:
        segments["low_value"].append(customer["name"])

print(segments)
```

### Output

```python
{
    'high_value': ['Asha', 'Karan'],
    'medium_value': ['Meena'],
    'low_value': ['Ravi']
}

# Task 3 — Session Deduplication

```python
events = [
    {"user_id": "U1", "page": "/home", "ts": 100},
    {"user_id": "U1", "page": "/home", "ts": 100},
    {"user_id": "U2", "page": "/cart", "ts": 101},
    {"user_id": "U1", "page": "/home", "ts": 105},
    {"user_id": "U2", "page": "/cart", "ts": 101},
]

seen = set()
unique_events = []
duplicates_removed = 0

for event in events:
    key = (event["user_id"], event["page"], event["ts"])

    if key not in seen:
        seen.add(key)
        unique_events.append(event)
    else:
        duplicates_removed += 1

print("Unique Events:")
print(unique_events)

print("Duplicates Removed:", duplicates_removed)
```

### Output

```python
Unique Events:
[
 {'user_id': 'U1', 'page': '/home', 'ts': 100},
 {'user_id': 'U2', 'page': '/cart', 'ts': 101},
 {'user_id': 'U1', 'page': '/home', 'ts': 105}
]

Duplicates Removed: 2
```

# Refactor 2 — O(n²) → O(n) Using a Set

### Refactored

````python
def has_duplicate(nums):
    seen = set()

    for num in nums:
        if num in seen:
            return True
        seen.add(num)

    return False

print(has_duplicate([1, 2, 3, 4]))
print(has_duplicate([1, 2, 3, 2]))
````

# Refactor 3 — Use collections.Counter

### Original
```python
counts = {}
for tag in ["bug", "feature", "bug", "bug", "chore", "feature"]:
    if tag in counts:
        counts[tag] += 1
    else:
        counts[tag] = 1
```
### Refactored

````python

from collections import Counter

counts = Counter(["bug", "feature", "bug", "bug", "chore", "feature"])

print(counts)
````

# Refactor 4 — Handle Empty Input and Missing Keys

### Refactored

````python
def average_score(students):
    scores = [s.get("score") for s in students if "score" in s]

    if not scores:
        return 0

    return sum(scores) / len(scores)

print(average_score([]))
print(average_score([{"name": "Asha"}]))
print(average_score([
    {"name": "Asha", "score": 90},
    {"name": "Ravi", "score": 80}
]))
````
