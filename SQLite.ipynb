# Basic Commands 

```python
#importing Libraries
import sqlite3 as sql
import pandas as pd
```

```python
# connect to a db (create if it doesnot exist)
connection = sql.connect('demo.db')
```

```python
# create a cursor object using connection
cursor = connection.cursor()
#cursor help you to traverse over the records in db
```

```python
#creating a table using cursor
cursor.execute('''CREATE TABLE people(
    id INTEGER PRIMARY KEY, 
    name TEXT NOT NULL,
    age INTEGER,
    city TEXT
);''')
```

```python
connection.commit()
```

```python
#commit everytime after creating or updating
#use various sql commands to manipulate
```

```python
cursor.execute('''INSERT INTO people(name, age,city) values ("Dev" ,32 , "Ambala");''')
cursor.execute('''Update people set name="sonam", age= 12 , city='Ladwa' where id =2;''')
```

```python
cursor.execute(''' SELECT * FROM people; ''')
```

```python
#To print data from db use fetchall() and print through a loop.
```

```python
results = cursor.fetchall()
```

```python
for record in results:
    print(record)
```

```python
df = pd.DataFrame(results)
```

```python
df
```

✨ Generating Synthetic Data ✨
</div>

```python
import pandas as pd
import random

first_names = ["Aarav", "Isha", "Kabir", "Meera", "Rohan"]
last_names = ["Sharma", "Verma", "Patel", "Khan", "Singh"]
cities = ["Delhi", "Mumbai", "Bangalore", "Chandigarh", "Jaipur"]

data = []

for _ in range(100):
    name = f"{random.choice(first_names)} {random.choice(last_names)}"
    age = random.randint(18, 60)
    city = random.choice(cities)

    data.append({
        "Name": name,
        "Age": age,
        "City": city
    })

df = pd.DataFrame(data)

print(df)
```

✨ Creating Connection with Database ✨
</div>

```python
conn = sql.connect("students.db")
```

```python
cursor = conn.cursor()
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS students (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        name TEXT,
        city TEXT,
        age INTEGER
    )
''')
```

```python
conn.commit()
```

✨ Converting DataFrame into Tuples and Inserting in Table ✨

```python
student_data = list(df.itertuples(index=False, name=None))

cursor.executemany("""
    INSERT INTO students (Name, City, Age)
    VALUES (?, ?, ?)
""", student_data)

conn.commit()
```
